```python
import pandas as pd
```

```python
cols = ['FECHA_INGRESO', 'FECHAALTA', "FECHA_NACIMIENTO"]
datos2022 = pd.read_csv("../data/GRD_PUBLICO_EXTERNO_2022.txt", encoding="utf-16", sep="|", engine="python", on_bad_lines="skip", usecols=cols)
datos2023 = pd.read_csv("../data/GRD_PUBLICO_2023.txt", encoding="utf-16", sep="|", engine="python", on_bad_lines="skip", usecols=cols)
datos2024 = pd.read_csv("../data/GRD_PUBLICO_2024.txt", encoding="latin-1", sep="|", engine="python", on_bad_lines="skip", usecols=cols)
```

```python
datos2022["FECHA_INGRESO"] = pd.to_datetime(datos2022["FECHA_INGRESO"])
datos2022["FECHAALTA"] = pd.to_datetime(datos2022["FECHAALTA"])
datos2022["FECHA_NACIMIENTO"] = pd.to_datetime(datos2022["FECHA_NACIMIENTO"], errors="coerce")

datos2022["DIAS"] = (datos2022["FECHAALTA"] - datos2022["FECHA_INGRESO"]).dt.days
datos2022["EDAD"] = (
    (datos2022["FECHA_INGRESO"] - datos2022["FECHA_NACIMIENTO"])
    .dt.days // 365
)
casos_raros = datos2022[(datos2022["EDAD"] == 0) & (datos2022["DIAS"] > 100)]
print(len(casos_raros))

# =========================
# 2023 (dayfirst=True)
# =========================
datos2023["FECHA_INGRESO"] = pd.to_datetime(datos2023["FECHA_INGRESO"], dayfirst=True)
datos2023["FECHAALTA"] = pd.to_datetime(datos2023["FECHAALTA"], dayfirst=True)
datos2023["FECHA_NACIMIENTO"] = pd.to_datetime(datos2023["FECHA_NACIMIENTO"], errors="coerce")

datos2023["DIAS"] = (datos2023["FECHAALTA"] - datos2023["FECHA_INGRESO"]).dt.days
datos2023["EDAD"] = (
    (datos2023["FECHA_INGRESO"] - datos2023["FECHA_NACIMIENTO"])
    .dt.days // 365
)
casos_raros = datos2023[(datos2023["EDAD"] == 0) & (datos2023["DIAS"] > 100)]
print(len(casos_raros))
# =========================
# 2024 (con "DESCONOCIDO")
# =========================
datos2024["FECHA_INGRESO"] = pd.to_datetime(datos2024["FECHA_INGRESO"], dayfirst=True, errors="coerce")
datos2024["FECHAALTA"] = pd.to_datetime(datos2024["FECHAALTA"], dayfirst=True, errors="coerce")
datos2024["FECHA_NACIMIENTO"] = pd.to_datetime(datos2024["FECHA_NACIMIENTO"], errors="coerce")

datos2024 = datos2024.dropna(subset=["FECHA_INGRESO", "FECHAALTA"])

datos2024["DIAS"] = (datos2024["FECHAALTA"] - datos2024["FECHA_INGRESO"]).dt.days
datos2024["EDAD"] = (
    (datos2024["FECHA_INGRESO"] - datos2024["FECHA_NACIMIENTO"])
    .dt.days // 365
)
casos_raros = datos2024[(datos2024["EDAD"] == 0) & (datos2024["DIAS"] > 100)]
print(len(casos_raros))

# Contar desconocidos (ahora son NaT)
desconocidos_ingreso_2024 = datos2024["FECHA_INGRESO"].isna().sum()
desconocidos_alta_2024 = datos2024["FECHAALTA"].isna().sum()
```

```python
fecha_corte = pd.Timestamp("2022-01-01")

datos2022 = datos2022[datos2022["FECHA_INGRESO"] >= fecha_corte]
datos2023 = datos2023[datos2023["FECHA_INGRESO"] >= fecha_corte]
datos2024 = datos2024[datos2024["FECHA_INGRESO"] >= fecha_corte]
```

```python
dias = 200
# 2022
mas_100_2022 = (datos2022["DIAS"] > dias).sum()
df_mas_100_2022 = datos2022[datos2022["DIAS"] > dias]

# 2023
mas_100_2023 = (datos2023["DIAS"] > dias).sum()
df_mas_100_2023 = datos2023[datos2023["DIAS"] > dias]

# 2024
mas_100_2024 = (datos2024["DIAS"] > dias).sum()
df_mas_100_2024 = datos2024[datos2024["DIAS"] > dias]

print("2022:", mas_100_2022)

print("2023:", mas_100_2023)

print("2024:", mas_100_2024)
```
