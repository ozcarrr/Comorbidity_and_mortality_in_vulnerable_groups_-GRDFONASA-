```python
import pandas as pd


datos2022 = pd.read_csv("../data/GRD_PUBLICO_EXTERNO_2022.txt", encoding="utf-16", sep="|", engine="python", on_bad_lines="skip")
datos2023 = pd.read_csv("../data/GRD_PUBLICO_2023.txt", encoding="utf-16", sep="|", engine="python", on_bad_lines="skip")
datos2024 = pd.read_csv("../data/GRD_PUBLICO_2024.txt", encoding="latin-1", sep="|", engine="python", on_bad_lines="skip")
```

```python
# Ver valores nulos
count = (datos2022['CIP_ENCRIPTADO'] == "DESCONOCIDO").sum()
print(count)
count = (datos2023['CIP_ENCRIPTADO'] == "DESCONOCIDO").sum()
print(count)
count = (datos2024['ID_BENEFICIARIO'] == "DESCONOCIDO").sum()
print(count)
```

```python
datos2024 = datos2024.rename(columns={"ID_BENEFICIARIO": "CIP_ENCRIPTADO"})
data_fechas = [col for col in datos2023.columns if "fecha" in col.lower()]
for x in data_fechas:
    datos2022[f"{x}"] = pd.to_datetime(datos2022[f"{x}"], dayfirst=True, errors="coerce")
    datos2023[f"{x}"] = pd.to_datetime(datos2023[f"{x}"], dayfirst=True, errors="coerce")
    datos2024[f"{x}"] = pd.to_datetime(datos2024[f"{x}"], dayfirst=True, errors="coerce")
cols1 = set(datos2022.columns)
cols2 = set(datos2023.columns)
cols3 = set(datos2024.columns)

diferentes = (cols1 | cols2 | cols3) - (cols1 & cols2 & cols3)

print(diferentes)
print("Solo en df1:", cols1 - cols2 - cols3)
print("Solo en df2:", cols2 - cols1 - cols3)
print("Solo en df3:", cols3 - cols1 - cols2)

print("df1 vs df2:", cols1.symmetric_difference(cols2))
print("df1 vs df3:", cols1.symmetric_difference(cols3))
print("df2 vs df3:", cols2.symmetric_difference(cols3))
```

```python
def comparar_columnas(*dfs):
    columnas = set()
    for df in dfs:
        columnas.update(df.columns)

    resultado = {}

    for i, df in enumerate(dfs):
        ejemplos = []
        for col in columnas:
            if col in df.columns:
                serie = df[col].dropna()
                valor = serie.iloc[0] if not serie.empty else None
                ejemplos.append(f"{valor} ({df[col].dtype})")
            else:
                ejemplos.append("NO EXISTE")
        resultado[f"df{i+1}"] = ejemplos

    return pd.DataFrame(resultado, index=list(columnas))

muestra = comparar_columnas(datos2022, datos2023, datos2024)
print(muestra)
```
